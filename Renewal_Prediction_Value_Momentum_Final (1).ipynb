```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, r2_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Step 2: Load Dataset
df = pd.read_csv("Insurance_dataSet.csv")

# Step 3: Feature Engineering
df['Claim_ratio'] = df['total_claim_amount'] / df['policy_annual_premium']
df['Interaction_score'] = (
    df['authorities_contacted'].apply(lambda x: 1 if x == 'Yes' else 0) +
    df['witnesses'].fillna(0) +
    df['police_report_available'].apply(lambda x: 1 if x == 'Yes' else 0)
)
df['Renewed'] = np.where((df['Seniority'] > 2) & (df['policy_annual_premium'] > 500), 1, 0)

# Step 4: Date-Based Features
df['Date_birth'] = pd.to_datetime(df['Date_birth'])
df['Date_driving_licence'] = pd.to_datetime(df['Date_driving_licence'])
df['Date_start_contract'] = pd.to_datetime(df['Date_start_contract'])

df['Customer_age'] = (pd.to_datetime('today') - df['Date_birth']).dt.days // 365
df['Driving_experience'] = (pd.to_datetime('today') - df['Date_driving_licence']).dt.days // 365
df['Policy_tenure'] = (pd.to_datetime('today') - df['Date_start_contract']).dt.days // 365

# Simulate Payment_history based on policy_annual_premium
df['Payment_history'] = np.where(df['policy_annual_premium'] > 1000, 'On-time',
                          np.where(df['policy_annual_premium'] > 500, 'Late', 'Missed'))

# Step 5: Label Encode Payment_history
le = LabelEncoder()
df['Payment_history_encoded'] = le.fit_transform(df['Payment_history'])

# Step 6: Select Features and Target
X = df[[
    'policy_annual_premium', 'total_claim_amount', 'Interaction_score',
    'Seniority', 'Policies_in_force', 'umbrella_limit',
    'Customer_age', 'Policy_tenure', 'Driving_experience', 'Claim_ratio',
    'Payment_history_encoded'
]]

y = df['Renewed']

# Step 7: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 8: Initialize and Train Models
model_lr = LogisticRegression(max_iter=500)
model_rf = RandomForestClassifier(n_estimators=200, random_state=42)
model_xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

model_lr.fit(X_train, y_train)
model_rf.fit(X_train, y_train)
model_xgb.fit(X_train, y_train)

# Step 9: Predictions
y_pred_lr = model_lr.predict(X_test)
y_pred_rf = model_rf.predict(X_test)
y_pred_xgb = model_xgb.predict(X_test)

# Step 10: Evaluation Metrics
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'Accuracy': [accuracy_score(y_test, y_pred_lr), accuracy_score(y_test, y_pred_rf), accuracy_score(y_test, y_pred_xgb)],
    'ROC_AUC': [roc_auc_score(y_test, y_pred_lr), roc_auc_score(y_test, y_pred_rf), roc_auc_score(y_test, y_pred_xgb)],
    'R2_Score': [r2_score(y_test, y_pred_lr), r2_score(y_test, y_pred_rf), r2_score(y_test, y_pred_xgb)]
})
print("🔹 Model Comparison:\n", results)

print("\n📊 Logistic Regression Report:\n", classification_report(y_test, y_pred_lr))
print("\n📊 Random Forest Report:\n", classification_report(y_test, y_pred_rf))
print("\n📊 XGBoost Report:\n", classification_report(y_test, y_pred_xgb))

# Step 11: Predict Renewal Probabilities with Best Model (XGBoost)
df['Renewal_Probability'] = model_xgb.predict_proba(X)[:, 1]

# Step 12: Segment Customers by Risk
df['Renewal_Risk'] = pd.cut(df['Renewal_Probability'],
                            bins=[0, 0.4, 0.7, 1.0],
                            labels=['High Risk', 'Medium Risk', 'Low Risk'])

# Step 13: Feature Importance (Random Forest)
importances = model_rf.feature_importances_
importance_percent = 100 * importances / importances.sum()
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance (%)': importance_percent
}).sort_values(by='Importance (%)', ascending=False)

print("\n🔝 Top Feature Importances (%):\n", importance_df.head(10))

# Optional: Plot Feature Importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'][:10], importance_df['Importance (%)'][:10])
plt.gca().invert_yaxis()
plt.title("Top 10 Feature Importances")
plt.xlabel("Importance (%)")
plt.show()

# Step 14: Save Final Dataset
df.to_csv("Insurance_dataSet_with_predictions.csv", index=False)
```

# XGBoost Model Selectio Finally due to Good Accuracy with data set 

```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Step 2: Load Dataset
df = pd.read_csv("Insurance_dataSet.csv")

# Step 3: Feature Engineering
df['Claim_ratio'] = df['total_claim_amount'] / df['policy_annual_premium']
df['Interaction_score'] = (
    df['authorities_contacted'].apply(lambda x: 1 if x == 'Yes' else 0) +
    df['witnesses'].fillna(0) +
    df['police_report_available'].apply(lambda x: 1 if x == 'Yes' else 0)
)
df['Renewed'] = np.where((df['Seniority'] > 2) & (df['policy_annual_premium'] > 500), 1, 0)

# Step 4: Date-Based Features
df['Date_birth'] = pd.to_datetime(df['Date_birth'])
df['Date_driving_licence'] = pd.to_datetime(df['Date_driving_licence'])
df['Date_start_contract'] = pd.to_datetime(df['Date_start_contract'])

df['Customer_age'] = (pd.to_datetime('today') - df['Date_birth']).dt.days // 365
df['Driving_experience'] = (pd.to_datetime('today') - df['Date_driving_licence']).dt.days // 365
df['Policy_tenure'] = (pd.to_datetime('today') - df['Date_start_contract']).dt.days // 365

# Step 5: Simulate and Encode Payment_history
df['Payment_history'] = np.where(df['policy_annual_premium'] > 1000, 'On-time',
                          np.where(df['policy_annual_premium'] > 500, 'Late', 'Missed'))
le = LabelEncoder()
df['Payment_history_encoded'] = le.fit_transform(df['Payment_history'])

# Step 6: Select Features and Target
X = df[[
    'policy_annual_premium', 'total_claim_amount', 'Interaction_score',
    'Seniority', 'Policies_in_force', 'umbrella_limit',
    'Customer_age', 'Policy_tenure', 'Driving_experience', 'Claim_ratio',
    'Payment_history_encoded'
]]
y = df['Renewed']

# Step 7: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 8: Train XGBoost Model
model_xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model_xgb.fit(X_train, y_train)

# Step 9: Predictions and Evaluation
y_pred = model_xgb.predict(X_test)
y_proba = model_xgb.predict_proba(X_test)[:, 1]

print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("📊 ROC-AUC Score:", roc_auc_score(y_test, y_pred))
print("📋 Classification Report:\n", classification_report(y_test, y_pred))

# Step 10: Predict Renewal Probabilities for All Customers
df['Renewal_Probability'] = model_xgb.predict_proba(X)[:, 1]

# Step 11: Segment Customers by Risk
df['Renewal_Risk'] = pd.cut(df['Renewal_Probability'],
                            bins=[0, 0.4, 0.7, 1.0],
                            labels=['High Risk', 'Medium Risk', 'Low Risk'])

# Step 12: Save Final Dataset
df.to_csv("Insurance_dataSet_with_XGBoost_predictions.csv", index=False)
```

```python
import seaborn as sns

# Compare predictions vs actuals
comparison_df = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred,
    'Probability': y_proba
})
print("🔍 Prediction Comparison:\n", comparison_df.head(10))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Misclassified samples
misclassified = comparison_df[comparison_df['Actual'] != comparison_df['Predicted']]
print(" Misclassified Samples:\n", misclassified.head())
```
